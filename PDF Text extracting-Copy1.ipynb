```python
list2 = []
for i in list1:
    if i != "":
        list2.append(i)
```

```python
print(list2)
```

```python
dict_resolucion = { "RESOLUCION": [], "CONSIDERACIONES": [], "RESOLUCIONES": [], "DISPOSICIONES": [] }
```

```python
dict_resolucion
```

```python
resolucion = False
consideracion = False
resuelve = False
disposiciones = False

count_resolucion = 0
count_consideracion = 0
count_resuelve = 0
count_disposiciones = 0
lent_total = len(list2)
```

```python
for index, i in enumerate(list2):
    #print(i," ")
    #print(index, " ")
    if "RESOLUCIÓN" in i:
        count_resolucion = index
    elif "CONSIDERANDO" in i:
        count_consideracion = index
    elif "RESUELVE" in i:
        count_resuelve = index
    elif "DISPOSICIONES" in i:
        count_disposiciones = index
        
print(count_resolucion, " ",count_consideracion, " ",count_resuelve, " ",count_disposiciones, " ")
```

```python
for i in range(count_resolucion +1, count_consideracion):
    dict_resolucion["RESOLUCION"].append(list2[i])
```

```python
for i in range(count_consideracion+1, count_resuelve):
    dict_resolucion["CONSIDERACIONES"].append(list2[i])
```

```python
for i in range(count_resuelve+1, count_disposiciones):
    dict_resolucion["RESOLUCIONES"].append(list2[i])
```

```python
for i in range(count_disposiciones+1, lent_total):
    dict_resolucion["DISPOSICIONES"].append(list2[i])
```
